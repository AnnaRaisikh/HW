```python

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import seaborn as sns
import re
import io
import scipy.stats as stats
```

# 1 Диагностические тесты (RESET-тест)
1.1 sleep equation #1

$$H_0: \beta_1=\beta_2=\beta_3=0$$
$$H_1: \beta_1^2+\beta_2^2+\beta_3^2>0$$

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
model=smf.ols(data=df, formula='sleep~totwrk+age+male+smsa+south').fit()
hat_y=model.predict()
```

```python
hat_y
```

```python
df['out2']=hat_y**2
df['out3']=hat_y**3
df['out4']=hat_y**4
df
```

```python
model2=smf.ols(data=df, formula='sleep~totwrk+age+male+smsa+south+out2+out3+out4').fit()
```

```python
R2_un=model2.rsquared
R2_re=model.rsquared
```

```python
F_obs=(R2_un-R2_re)/(1-R2_un)*(len(df)-8)/3
F_obs
```

```python
alpha=0.05
stats.f.ppf(1-alpha,3,len(df)-8)
```

Fobs < Fcr => H0 не отвергаем, регрессия согласуется с данными

# 1.2 sleep equation #2

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
df['totwrk*south']=df['totwrk']*df['south']
df
```

```python
model1=smf.ols(data=df, formula='sleep~totwrk+south+totwrk*south+age+age**2+male+smsa+marr').fit()
ht_y=model1.predict()
ht_y
```

```python
df['slp2']=ht_y**2
df['slp3']=ht_y**3
df['slp4']=ht_y**4
df
```

```python
model2=smf.ols(data=df, formula='sleep~totwrk+south+totwrk*south+age+age**2+male+smsa+marr+slp2+slp3+slp4').fit()
```

```python
R2_un=model2.rsquared
R2_re=model1.rsquared
```

```python
F_obs=(R2_un-R2_re)/(1-R2_un)*(len(df)-12)/3
F_obs
```

```python
alpha=0.05
stats.f.ppf(1-alpha,3,len(df)-12)
```

Fobs < Fcr => H0 не отвергаем, регрессия согласуется с данными

# 1.4 output equation #2

```python
df = pd.read_csv('https://raw.githubusercontent.com/ryupepa/Econometrica/master/python-notebooks/data-csv/Labour.csv')
df
```

```python
model_4 = smf.ols(data=df, formula='np.log(output)~np.log(capital)+np.log(labour)+np.log(wage)').fit()
ht_y=model_4.predict()
ht_y
```

```python
df['logoutput2']=ht_y**2
df['logoutput3']=ht_y**3
df['logoutput4']=ht_y**4
df
```

```python
model5=smf.ols(data=df, formula='np.log(output)~np.log(capital)+np.log(labour)+np.log(wage)+logoutput2+logoutput3+logoutput4').fit()
```

```python
R2_un=model5.rsquared
R2_re=model_4.rsquared
```

```python
F_obs=(R2_un-R2_re)/(1-R2_un)*(len(df)-7)/3
F_obs
```

```python
alpha = 0.05
stats.f.ppf(1-alpha,3,len(df)-6).round(2)
```

Fcr < Fobs => H0 отвергаем, регрессия не согласуется с данными
